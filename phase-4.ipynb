Broadcasting in NumPy is basically automatic shape matching so you can do math between arrays of different shapes without writing loops.

```python
a = np.array([1, 2, 3])
b = 2
print(a + b)  # [3 4 5]
#here b is stretching it like a[1,2,3] +b[2,2,2] (this stretching is called broadcasting)

```

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
arr1 = np.array([1,2,3,4,5])
arr2 = np.random.rand(3,3)
arr3 = np.random.randint(5, size=(2, 4))
arr4 = np.zeros((4, 4))
````

````python
#save array we can also convert img into matrix
np.save('arr1.npy',arr1)
np.save("arr2.npy",arr2)
np.save("arr3.py",arr3)
np.save("arr4.npy",arr4)
````

````python
#load file and display
try:
    logo = np.load('numpy-logo.npy')

    # Display
    plt.figure(figsize=(10, 5))
    plt.subplot(121)
    plt.imshow(logo)
    plt.title("Numpy logo")
    plt.grid(False)

    dark_logo = 1 - logo

    plt.subplot(122)
    plt.imshow(dark_logo)
    plt.title("Numpy Dark logo")
    plt.grid(False)

except FileNotFoundError:
    print("numpy logo file not found")
````
